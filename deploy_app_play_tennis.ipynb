## Import Dataset

```python
#Untuk mengimport dataset

import pandas as pd
data = pd.read_csv("playTennis.csv")
data
```

## Drop Kolom

```python
#Untuk menghapus kolom day

data.drop('day', axis=1, inplace=True)
data
```

## Label Encoding

```python
#Untuk mengubah nilai suatu kolom menjadi angka terurut

from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['outlook'] = Le.fit_transform(data['outlook'])
data['temp'] = Le.fit_transform(data['temp'])
data['humidity'] = Le.fit_transform(data['humidity'])
data['wind'] = Le.fit_transform(data['wind'])
data['play'] = Le.fit_transform(data['play'])
data
```

## Atribut Kriteria

```python
#Mengelompokkan atribut yang digunakan sebagai kriteria
#Atribut outlook, temp, humidity, wind

x = data.iloc[:,:4].values
x
```

## Atribut Keputusan

```python
#Mengelompokkan atribut yang digunakan sebagai keputusan

y = data.iloc[:,-1].values
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.9)
```

## Klasifikasi Naive Bayes

```python
from sklearn import naive_bayes
from sklearn import metrics

nb = naive_bayes.BernoulliNB()
nb.fit(x_train, y_train)
y_pred = nb.predict(x_test)
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi prediksi: ", score)
```
